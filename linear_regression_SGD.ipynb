```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
from sklearn.datasets import make_regression
```

```python
data = make_regression(n_samples=100, n_features=1, bias=0.1, noise=25)
```

```python
x = data[0]
y = data[1]
```

```python
plt.scatter(x,y)
```

```python
def sgd(learning_rate, epochs=1000):
    
    m = 0
    c = 0
    n = float(len(x))
    
    for i in range(epochs):
        for j in range(len(x)):
            y_pred = m*x[j] + c
            derivative_m = (-2/n)* sum(x[j]*(y[j]-y_pred))
            derivative_c = (-2/n)* sum(y[j]-y_pred)

            m = m - learning_rate* derivative_m
            c = c - learning_rate* derivative_c
                
    return (m , c)
```

```python
theta1, theta2 = sgd(0.1, 1000)
```

```python
y_pred = theta1*x + theta2
```

```python
plt.scatter(x,y)
plt.plot([min(x),max(x)],[min(y_pred),max(y_pred)], color='red')
plt.show()
```
